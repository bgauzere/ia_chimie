# Random Forests

## Decision Tree
Learn how to separate the data

```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(centers=[[-1,-1],[1,1]], cluster_std=0.75,random_state=0)
colors = ['tab:green','tab:blue']
fig,ax = plt.subplots()

for k,classe in enumerate(np.unique(y)):
    ax.plot(X[y==classe,0],X[y==classe,1],'o',c=colors[k],markersize="10")

for spine in ['right','top']:
    ax.spines[spine].set_color("None")
#ax.set_axis_off()
plt.tight_layout()
plt.savefig("decision_tree_data.svg")
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X,y)
ypred = clf.predict(X)
```

```python
from sklearn import tree
tree.plot_tree(clf,label=None,filled=True,impurity=False);
plt.tight_layout()
plt.savefig("decision_tree_tree.svg")
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib 
fig,ax = plt.subplots()

colors = ['tab:green','tab:blue']

DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
    plot_method="contourf",
        ax=ax,
    alpha = .33,
    cmap=matplotlib.colors.ListedColormap(colors)
    )
for k,classe in enumerate(np.unique(y)):
    ax.plot(X[y==classe,0],X[y==classe,1],'o',c=colors[k],markersize="5")

for spine in ['top','right']:
    ax.spines[spine].set_color("None")
ax.grid()
plt.tight_layout()
plt.savefig("decision_tree_boundary.svg")
```

```python
from sklearn.ensemble import RandomForestClassifier
n_estimators = 20 # the number of trees in the forest
max_depth = None # expand as you can
max_features = "sqrt" # RTFM
clf = RandomForestClassifier(n_estimators=n_estimators,
                             max_depth=max_depth,
                             max_features=max_features)
clf.fit(X,y)
ypred = clf.predict(X)
```

```python
fig,ax = plt.subplots()

colors = ['tab:green','tab:blue']

DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
    plot_method="contourf",
        ax=ax,
    alpha = .33,
    cmap=matplotlib.colors.ListedColormap(colors)
    )
for k,classe in enumerate(np.unique(y)):
    ax.plot(X[y==classe,0],X[y==classe,1],'o',c=colors[k],markersize="5")

for spine in ['top','right']:
    ax.spines[spine].set_color("None")
ax.grid()
plt.tight_layout()
```

# Entropy


```python
import scienceplots

def entropy_measure(p):
    return - p * np.log2(p) - (1-p)*np.log2(1-p)

# Utiliser un style moderne
plt.style.use('science')
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Computer Modern Serif"
})
# Crer une couleur personnalise pour le graphique
palette = plt.get_cmap('Set1')
nb_c1 = np.arange(1,50)
nb_c2 = 50-nb_c1
p = nb_c1/50

entropy = entropy_measure(p)

labels = [f'C1={i}, C2={50-i}' for i in nb_c1]
fig,ax = plt.subplots(figsize=(5,3))
# Plot the data
ax.plot(p, entropy,label="entropy", color=palette(1), linewidth=2.5, alpha=0.9)


# Supprimer les pines du haut et de la droite
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Configurer l'axe des y pour commencer  0
ax.set_ylim(0, max(entropy) + .1)

# Ajouter une lgende

# Dfinir un intervalle pour les major ticks (par exemple, tous les 5)
major_ticks = p[::5]
major_labels = labels[::5]

# Configurer les major ticks et leurs tiquettes
ax.set_xticks(major_ticks)
ax.set_xticklabels(major_labels, rotation=45)

ax.set_xlabel("Class repartition")
ax.set_ylabel("Entropy")

# Display the plot
plt.tight_layout()  # To ensure labels don't overlap
plt.savefig("entropy.svg")
plt.show()
```

```python
def ce_measure(p):
    # vraie def    return 1 - max(p-1)
    #2 class def
    return [min(pi,1-pi) for pi in p]
    

import scienceplots
plt.style.use('science')
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Computer Modern Serif"
})
palette = plt.get_cmap('Set1')
nb_c1 = np.arange(1,50)
nb_c2 = 50-nb_c1
p = nb_c1/50

ce = ce_measure(p)


labels = [f'C1={i}, C2={50-i}' for i in nb_c1]
fig,ax = plt.subplots(figsize=(5,3))
# Plot the data
ax.plot(p, ce,label="Classification error", color=palette(1), linewidth=2.5, alpha=0.9)


# Supprimer les pines du haut et de la droite
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Configurer l'axe des y pour commencer  0
ax.set_ylim(0, max(ce) + .1)

# Ajouter une lgende

# Dfinir un intervalle pour les major ticks (par exemple, tous les 5)
major_ticks = p[::5]
major_labels = labels[::5]

# Configurer les major ticks et leurs tiquettes
ax.set_xticks(major_ticks)
ax.set_xticklabels(major_labels, rotation=45)

ax.set_xlabel("Class repartition")
ax.set_ylabel("Classification error")

# Display the plot
plt.tight_layout()  # To ensure labels don't overlap
plt.savefig("ce.svg")
plt.show()

```
